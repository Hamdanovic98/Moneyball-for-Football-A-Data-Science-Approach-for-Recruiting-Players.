```python
import requests

import pandas as pd

# Define your column names
columns = [
    'Player', 'Nation', 'Pos', 'Squad', 'Comp', 'Age', 'Born', 'MP', 'Starts', 'Min',
    '90s', 'Goals', 'Assists', 'G+A', 'G-PK', 'PK', 'PK_Attempted', 'Yellow', 'Red', 'xG', 'npxG',
    'xAG', 'npxG+xAG', 'Prg_Carries', 'Prg_Passes', 'Prg_Passes_Received', 'Goals_per90', 'Assits_per90', 'G+A_per90', 'G-PK_per90',
    'G+A-PK_per90', 'xG_per90', 'xAG_per90', 'xG+xAG_per90', 'npxG_per90', 'npxG+xAG_per90','Matches'
]

# Create an empty dataframe with those columns
df = pd.DataFrame(columns=columns)

# Specify the URL of the page
url = "https://fbref.com/en/comps/Big5/stats/players/Big-5-European-Leagues-Stats"

# Send HTTP request to URL
response = requests.get(url)

# Parse HTML and save to BeautifulSoup object
soup = BeautifulSoup(response.text, "html.parser")

# Find table body
table_body = soup.find("tbody")

# Find rows in the table body
rows = table_body.find_all("tr")

# Parse rows
for row in rows:
    cols = row.find_all('td')
    if cols:
        data = [col.get_text() for col in cols]
        if len(data) != len(columns):
            print("Data: ", data)
        else:
            df_length = len(df)
            df.loc[df_length] = data

df
```

```python
df.to_csv('2023/Players_Standard_2023.csv')
```
