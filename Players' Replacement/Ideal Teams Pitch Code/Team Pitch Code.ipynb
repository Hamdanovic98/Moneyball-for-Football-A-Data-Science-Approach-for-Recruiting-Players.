```python
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import mplsoccer
from mplsoccer import VerticalPitch

# Load the font
font_path = '../../Fonts/Ubuntu/Ubuntu-Medium.ttf'
font_prop = fm.FontProperties(fname=font_path, size=15)  # Adjust size as needed

# Initialize pitch with a smaller size
pitch = VerticalPitch(pitch_type='opta', pitch_color='black', line_color='white')
fig, ax = pitch.draw(figsize=(8, 12))  # Reduced figure size

# Define positions (x, y) for each player in a 4-2-3-1 formation
positions = {
    'Stegen': (5, 50),  # Goalkeeper
    'Bald': (20, 90),  # Left Back
    'Iigo Martinez': (20, 65),  # Left Center Back
    'Kound': (20, 35),  # Right Center Back
    'Santos': (20, 10),  # Right Back
    'De Jong': (40, 65),  # Left Pivot
    'Christensen': (40, 35),  # Right Pivot
    'Olmo': (70, 50),  # Attacking Midfielder (Number 10)
    'Fhrich': (70, 85),  # Left Winger
    'Yamal': (70, 15),  # Right Winger
    'David': (78, 50),  # Striker (Number 9)
}

# Annotate player names on the pitch
for player, pos in positions.items():
    pitch.annotate(player, xy=pos, va='center', ha='center', color='white', fontsize=15, ax=ax, fontproperties=font_prop)

# Title with custom font and padding
ax.set_title('Barcelona 2024/25: Rebuilding the Squad the Data Way', color='white', fontproperties=font_prop, fontsize=18, pad=20)

# Set facecolor to black to ensure the background matches
fig.patch.set_facecolor('black')
ax.set_facecolor('black')

# Save the figure with tight bounding box to remove extra borders
plt.savefig('Barcelona_2025.png', dpi=500, bbox_inches='tight', facecolor='black', edgecolor='none')

# Show plot
plt.show()
```
